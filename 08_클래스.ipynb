## 1.1 클래스의 기본구조

클래스는 하나의 개체 안에 함수와 변수를 넣어 유기적으로 사용할 수 있게 만드는 것입니다.

함수와 변수를 하나로 묶는다?   
마치 저번에 소개한 모듈과 같죠? 하지만 그것보다 더 많은 기능과 특별함이 있습니다.

클래스는 함수 같기도 하고 모듈 같기도 하고 또 데이터 같기도 합니다... 신기하죠?

클래스의 구조는 다음와 같습니다.

image.png

````python
class 클래스 이름:
    def __init__(self, 인자): #생성자, 초기화함수(변수 선언)
        변수 초기화

    def 함수(self, 인자):
        코드

인스턴스 = 클래스()
````

자 복잡하게 생각하기 보다 바로 예제를 하나 만들어 봅시다.

````python
class Calculator:
    def __init__(self, init_value = 0):         #init은 클래스가 처음 생성될 때 자동으로 실행됩니다.
        self.value = init_value                 #init안에 변수들을 초기화 할 수 있습니다.

    def add(self, num):         #클래스 내부의 함수를 메서드(method)라고 부릅니다.
        self.value += num      #method 내부에서 클래스의 변수를 사용할 때에는 self로 접근합니다.
        return self.value

    def sub(self, num):         #메서드의 처음 parameter는 항상 self로 시작합니다.
        self.value -= num
        return self.value


calc = Calculator(7)             #생성(선언) = 호출, 클래스는 사용하기 위해서는 반드시 인스턴스(instance, 객체)를 만들어야 합니다.
calc.__init__(7)
print('value', calc.value)    #인스턴스 내부의 변수(어트리뷰트)에 접근하려면 . 을 사용하면 됩니다.
print('add', calc.add(10))      #인스턴스 내부의 함수(메서드)도 마찬가지 입니다.
````

## 1.2 self란?

객체의 대명사.

self는 현재 인스턴스를 가리키는 매개변수로, 해당 메서드를 호출한 객체 자신을 참조하는 역할을 합니다.    
self를 통해 객체 내부의 속성(attribute)에 접근하고, 메서드를 호출할 수 있습니다.

메서드의 첫번째 파라미터는 반드시 self여야 합니다.   
(예외는 나중에 배웁시다.)

메서드를 사용할 때에는 self를 제외한 뒤에 argument만 넣어주면 됩니다.

image.png

````python
class Person:
    def __init__(self, name):
        self.name = name
        #인자로 받는 name과 self.name은 다릅니다. 우연히 같은거에요!!

    def say_hello(self):
        print(f"안녕하세요, 저는 {self.name}입니다.")
        #self를 사용하면 메서드가 달라도 어디서나 변수를 사용하거나 바꿀 수 있습니다.

# Person 클래스의 인스턴스 생성
person = Person("John")
person.say_hello()
````

## 1.5 클래스의 변수들

앞서 인스턴스끼리는 서로 독립적이라 영향을 주지 않는다고 했었습니다.   
예외가 있는데 바로 클래스 변수입니다.  

클래스 변수는 클래스 바로 아래에 선언됩니다.

````python
class MyCls:
    cls_var = 0

    def __init__(self, a):
        self.inst_var = a

#두 인스턴스를 생성
c = MyCls(3)
d = MyCls(3)

#c의 인스턴스 변수를 변경
c.inst_var = 10

print('c의 인스턴스 변수 :', c.inst_var)
print('d의 인스턴스 변수 :', d.inst_var)   #c의 변수를 바꿔도 d의 값에는 영향을 주지 않는다.
print()

#MyCls의 클래스 변수를 변경
MyCls.cls_var = 100

print('c의 클래스 변수 :', c.cls_var)
print('d의 클래스 변수 :', d.cls_var)     #모든 인스턴스에 영향을 준다.
````

클래스 변수, 인스턴스 변수, 지역 변수는 선언되는 위치와 앞에 붙는 self에 따라 구분되니 신경써서 봅시다.

- class 변수 : def(메서드) 밖에 선언되며 하나의 class에서 생성된 모든 인스턴스에 영향을 줄 수 있다.
- instance 변수 : self(인스턴스)가 붙어있고 인스턴스 내 모든 위치에서 통용된다.
- local 변수 : 해당 메서드 내에서만 존재하며 밖에서는 사용할 수 없다.

image.png

````python
#method를 이용해서 클래스 변수를 바꾸는 법
class MyCls:
    cls_var = 0

    def __init__(self, a):
        self.inst_var = a

    def add_instance_var(self, var):
        self.inst_var += var

    def add_cls_var(self, var2):
        MyCls.cls_var += var2   #class 이름으로 클래스 변수에 접근할 수 있다.

c = MyCls(3)
d = MyCls(3)
c.add_instance_var(5)
print(c.inst_var, d.inst_var)

print(MyCls.cls_var)
MyCls.cls_var += 1
print(c.cls_var, d.cls_var)

c.add_cls_var(100)
print(c.cls_var, d.cls_var)
````

````python
class Person:
    people_num = 0

    def __init__(self, n):
        self.name = n
        Person.people_num += 1

    def __del__(self):
        Person.people_num -= 1

# 사람을 삭제하면 전체 사람 숫자도 줄어들게

a = Person('Tom')
print(a.people_num)

b = Person('Jack')
print(a.people_num, b.people_num)

c = Person('Jenny')
print(a.people_num, b.people_num, c.people_num)

del c
print(a.people_num, b.people_num)
````

## 1.6 method의 종류

### 클래스 내의 함수(메서드)의 종류
1) 인스턴스 함수(메서드): 객체(인스턴스)의 기능  
2) 클래스 함수(메서드): 클래스 내 관리  
- 클래스 함수는 첫번째 매개 변수로 클래스 인스턴스가 아닌 클래스 자체가 넘어오게 됩니다.    

3) 정적(스태틱) 함수(메서드): 사실 상 클래스와 무관/관리의 차원  
- 정적 메서드를 호출할 때는, 첫번째 인자로 아무것도 넘어오지 않기 때문에, 명시적으로 넘긴 다른 인자만 접근할 수 있습니다.

class method, static method는 아직은 크게 신경쓰지 않아도 됩니다.

````python
class Vehicle:

    instance_count = 0

    def __init__(self, wheel, color, model_code):
        self.wheel = wheel
        self.color = color
        self.model_code = model_code
        Vehicle.instance_count += 1

    #인스턴스 메서드(함수)
    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))

    #인스턴스 메서드(함수)
    def stop(self):
        print('정지!')

    #클래스 메서드(함수)
    @classmethod # decorator(장식자, 데코레이터) : 기존의 코드에 여러가지 기능을 추가하는 파이썬 구문
    def count_instance(cls):
        print('현재 인스턴스 수 :{}'.format(Vehicle.instance_count))

    @staticmethod #self 인자가 없음. 독립적인 함수
    def check_type(model_code):
        if model_code >=20:
            print('전기차입니다.')
        elif model_code >= 10:
            print('수소차입니다.')
        else:
            print('디젤차입니다.')
````

````python
mohabi = Vehicle(21, 'black', 5)
mohabi.move(400)
mohabi.check_type(mohabi.model_code)

sonata = Vehicle(19, 'white', 27)
Vehicle.count_instance()  #sonata.count_instance()
Vehicle.check_type(mohabi.model_code)
````

````python
#아래 함수들을 mymodel이라는 class로 선언할 것

class MyModel:

    def __init__(self):
        self.weight = 10

    def predict(self, x): #인스턴스 함수
        result = self.weight * x
        return result

    def evaluate(self, x, y):#인스턴스 함수
        pred = self.weight * x
        err = y- pred
        return err

    @staticmethod
    def check_model_type(model_code):#스태틱 함수
        if model_code > 10:
            return 'neuralnetwork'
        elif model_code > 0:
            return 'machinelearning'
        else:
            return 'etc'
````

````python
model = MyModel()
print(model.weight)

print(model.predict(10))

print(model.evaluate(10, 99))

print(model.check_model_type)
````

## 2.1 상속

기존의 클래스와 일부 겹치는 요소를 갖는 새로운 클래스를 만들 때, 겹치는 부분을 그대로 가져다 쓸 수 있으면 편하지 않을까?

상속은 클래스를 만들 때 개별로 작성하는게 아니라   
더 큰 개념의 공통된 클래스를 만들고 세부적인 특징은 상속받아 추가할 수 있습니다.

우선 상속을 활용하기 앞서 우리가 했던 방식으로 고양이, 강아지, 오리 3종류 동물을 만들어봅시다.   
각 동물은 자기를 소개하고 우는 소리를 알려줄 수 있고 다리의 개수를 속성으로 가지고 있습니다.

````python
class Cat:
    def __init__(self, name):
        self.name = name
        self.feet_num = 4

    def sound(self):
        print("야옹")

    def intro(self):
        print(f"Hi~ I'm {self.name}")

class Dog:
    def __init__(self, name):
        self.name = name
        self.feet_num = 4

    def sound(self):
        print("멍멍")

    def intro(self):
        print(f"Hi~ I'm {self.name}")

class Duck:
    def __init__(self, name):
        self.name = name
        self.feet_num = 2
        self.is_wings = True

    def sound(self):
        print("꽥꽥")

    def intro(self):
        print(f"Hi~ I'm {self.name}")
````

이것외에도 우리가 여러 종류의 동물(소, 치타, 닭, 타조, 펭귄, 호랑이, 사자)을 만든다고 생각해봅시다.

동물들은 서로 다른 특성도 있지만 서로 같은 기능이나 속성도 많이 있습니다. 우리가 하나하나 똑같이 만들어 줘야 할까요?

아래 코드는 상속을 사용하여 동물을 만든 전체적인 코드 입니다.   
천천히 살펴 봅시다.

````python
class Animal:
    feet_num = 4

    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f'...{self.name}은 아무 말이 없다.')

    def intro(self):
        print(f"Hi~ I'm {self.name}")

class Cat(Animal):
    def sound(self): #이름 일치 시키고 내가 쓰고 싶은 걸로 덮어씌움. 커스텀 = 매서드오버라이딩
        print("야옹")

class Giraffe(Animal):
    pass

class Dog(Animal):
    def sound(self):
        print("멍멍")

class Bird(Animal):
    feet_num = 2
    is_wings = True

class Duck(Bird):
    def sound(self):
        print("꽥꽥")
````

위의 코드는 개별 동물을 만들기에 앞서 모든 동물에 공통적인 기능과 속성을 포함하는 'Animal'이라는 클래스를 만들어 냅니다.

그리고 하위 동물들은 Animal이라는 클래스를 `상속`받아 동물들을 만들어 냅니다.    
개별 클래스에는 feetnum을 지정해주지 않아도 자동으로 feetnum은 4개로 상속받게 됩니다.

추가로 오리를 만들려고 했는데 오리는 앞서 만들었던 Animal들과 특징이 달라서    
Bird라는 클래스를 Animal로부터 상속받아 기본적인 특성을 수정합니다.

뼈대가 되는 상위 클래스를 parents, 상속받아 다시 만들어낸 하위 클래스를 child 클래스라고 합니다.

````python
dog = Dog('강아지')
cat = Cat('고양이')
duck = Duck('오리')

dog.sound()
cat.sound()
duck.sound()
````

````python
print(dog.feet_num)                 #Animal로부터 만든 클래스는 4개의 발을 가지고 있다.
print(cat.feet_num)
print(duck.feet_num, duck.is_wings) #Bird로부터 만든 클래스는 2개의 발과 날개를 가지고 있다.
````

## 2.2 메소드 오버라이딩 (overriding, 덮어쓰기)

앞선 동물 예제에서 sound를 살펴 봅시다.

왜 다시 sound를 만들어야 했을까?

### 상속과 오버라이딩 중간 정리

클래스 상속에서 "parent class"(또는 super class)는 기존에 정의된 클래스를 말하며,   
"child class"(또는 sub class)는 상속받은 새로운 클래스를 말합니다.

>Parent class는 상속을 통해 자신의 특성과 동작을 하위 클래스에게 전달합니다.   
Parent class는 기본적으로 독립적으로 사용될 수 있지만, 하위 클래스에 의해 확장되거나 변경될 수도 있습니다.

>Child class는 상속을 받은 클래스로, Parent class의 특성과 동작을 상속받아 사용하면서, 필요에 따라 새로운 속성과 메서드를 추가하거나 Parent class의 메서드를 재정의할 수 있습니다.
   
Child class는 Parent class의 모든 속성과 메서드를 상속받으며, 이를 통해 코드 재사용성과 구조적인 계층 구조를 구현할 수 있습니다.

객체 지향 프로그래밍에서는 코드를 모듈화하고 구조화하여 유지보수와 확장성을 향상시킬 수 있습니다.

>쉽게 생각하면 됩니다. 상속을 받을 때 공통된건 그대로 두고 다른건 새로 오버로딩 하면 됩니다.

## 2.2 super

앞서 잠깐 언급한대로 parent class는 super라고 불리기도 합니다.

child에서 super()는 내가 상속받은 상위 클래스에 바로 접근할 수 있게 합니다.

아래 예제를 보면 overriding을 조금 더 섬세하게 쓸 수 있습니다.

````python
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"안녕하세요, 저는 {self.name}입니다.")


class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   #super().__init__은 Parent.__init__과 같습니다.
        self.age = age

    def greet(self):
        super().greet()   #Parent.greet()와 동일합니다. 부모클래스 greet 직접적으로 호출하고 싶을 때,
        print(f"저의 나이는 {self.age}살입니다.")

        #기존의 greet를 완전히 덮에쓰는게 아니라 그대로 유지한 채 추가할 수 있습니다.

child = Child("철수", 10)
child.greet()
````

# 3. 매직 메소드(magic or special method)

클래스에 메서드를 만들어 넣을 수 있지만 사실 클래스에는 기본적으로 생성되는 많은 특수한 메소드들이 있습니다.

이런 메소드들은 좌우에 `__`가 붙어있는데 이런 메소드를 매직 메소드 혹은 스페셜 메소드라고 부릅니다.

매직 메소드는 정해진 오퍼레이터나 함수를 사용하면 자동으로 실행됩니다.

````python
#int에 들어있는 magic method를 열어봅시다.
for m in dir(int(3)):
    if m.startswith('__'):
        print(m)
````

````python
#사칙연산과 관련된 매직 매소드
a = 3
b = 4

print(a + b, a.__add__(b))
print(a - b, a.__sub__(b))
print(a * b, a.__mul__(b))
#등등
````

````python
#비교와 관련된 매직 메소드

print(a < b, a.__lt__(b))
print(a == b, a.__eq__(b))
print(a >= b, a.__ge__(b))
````

## 3.1. 매직 매소드를 이용한 나만의 데이터 타입 만들어보기

아래 예시들은 매직 매소드와 상속을 사용해 새로운 재미있는 나만의 데이터를 만드는 예시입니다.

우리가 배우는 수준을 넘어서기 때문에 참고용으로 보고 너무 어렵게 받아들이지 않아도 됩니다.^^

````python
# +와 -를 오버라이딩으로 바꾸기
# *를 str처럼 동작하게 하기

class BaBoint(int):
    def __add__(self, n):
        return super(BaBoint, self).__sub__(n)

    def __sub__(self, n):
        return super(BaBoint, self).__add__(n)

    def __mul__(self, n):
        return int(super(BaBoint, self).__str__() * n)
````

````python
a = BaBoint(7)
b = BaBoint(3)

print(a + b)
print(a - b)
print(a * 3)
````

````python
#숫자도 더할 수 있는 리스트 만들기
import numbers
class SuperList(list):

    def __add__(self, x):
        if isinstance(x, list):
            return super().__add__(x)
        elif isinstance(x, numbers.Number):
            return [i + x for i in self]

li = SuperList([1, 2, 3, 4])
print(li + [-1, -2])
print(li + 5)
````

````python
#짝수는 날려버리는 리스트
class OddList(list):
    def __init__(self, arg):
        odd = [i for i in arg if i%2 == 1]
        super().__init__(odd)

OddList([1, 2, 3, 4, 5])
````

````python
#자동으로 글자 뒤집
class MirrorStr(str):
    def __new__(cls, string):
        return super().__new__(cls, string[::-1])

MirrorStr('안녕하세요')
````
